```python
import gtts
import pandas as pd
from tqdm import tqdm
```

```python
drug_names = pd.read_excel('../data/Drug-Dosing.xlsx',engine='openpyxl')['Generic Name'].to_list()
```

```python
drug_names
```

```python
for drug in tqdm(drug_names):
    tts = gtts.gTTS(drug)
    tts.save(f'../data/recordings//{drug}.mp3')
```
